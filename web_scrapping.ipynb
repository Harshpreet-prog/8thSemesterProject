```python
import pandas as pd

ls = pd.read_csv("/content/drive/MyDrive/Amazon Scraping - Sheet1.csv")
ls.drop("id",axis = 1,inplace = True)
ls
```

```python
import time
import pandas as pd
from bs4 import BeautifulSoup
from requests_html import HTMLSession
import json

s = HTMLSession()

def get_data(url):
    webpage = s.get(url)
    soup = BeautifulSoup(webpage.content, "html.parser")
    try:
       product_title = soup.find("span",attrs={"id": 'productTitle'}).text
    
    except AttributeError:
      return "file not found"
    
    try:
       product_image_url =soup.find("img",{"alt": ""})['src']
   
    except AttributeError:

       return "file not found"
   
    try:

       product_price = (soup.find("span", class_="a-color-base").text)

    except AttributeError:

      return "file not found"
   
    try:
       product_details = []
       details = soup.findAll("span", {"class":"a-list-item"})
       for tags in details:
         value = tags.text.replace("\n", "")
         product_details.append("".join(value))

    except AttributeError:
      return "file not found"
   
    
    return [{"product_title" : product_title ,"product_image_url": product_image_url,"product_price" : product_price,
            "product_details": product_details}] 

def save_data(collectedData):
  with open('dataCollected.json', 'w') as myfile:
    try:
      json.dump(collectedData,myfile)
      print(collectedData)
    except TypeError:
      print("cant dump")
    

def main():
    start_time = time.time()
    print("saving the output of extracted information")
    url1 ="https://www.amazon."
    url2 = "/dp/"
    new_file = ls[["Asin","country"]]

    for index,row in new_file.iterrows():
        url = url1 + row["country"] + url2 + row["Asin"]
        #print(url)
        data =  get_data(url)
        save_data(data)
 
        #web scrapinf code here  
    
    time_diffrence = time.time() - start_time
    print("scrapping time: %2f secondfs" %time_diffrence)
     
main()

```
